```python
import numpy as np
import matplotlib.pyplot as plt

from mesa import Model, Agent
from mesa.time import RandomActivation
from mesa.space import MultiGrid, SingleGrid

import random
import itertools
```

```python
class Citizen(Agent):
    
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        race = random.randint(1,self.model.n_races)
        self.race = race
        
    def step(self):
        self.look_neighbor()
        self.similarity()
        self.decide()
        self.relocate()
        
    def look_neighbor(self):
        """search for neighbors"""
        self.neighbors = self.model.grid.get_neighbors(
            self.pos, moore = True, include_center = False)
        
    def similarity(self):
        """calculate race similarity"""
        if len(self.neighbors) == 0:
            self.satisfaction = 1.0
        else:
            vfunc = np.vectorize(lambda x: x.race)
            sim = self.race == vfunc(self.neighbors)
            diff = (sim == False).sum()
            same = (sim == True).sum()
            self.satisfaction = same/(same+diff)
            
    def decide(self):
        """decide to stay or move"""
        if self.satisfaction <= self.model.sim_threshold:
            self.move = True
        else:
            self.move = False
        
    def relocate(self):
        if self.move:
            self.model.grid.move_to_empty(self)

class City(Model):
    def __init__(self, width, height, n_citizens, n_races, sim_threshold):
        self.schedule = RandomActivation(self)
        self.n_steps = 0
        
        self.grid = SingleGrid(width, height, False)
        loc = list(itertools.product(range(width),range(height)))
        self.grid.grids = loc
        random.shuffle(loc)

        self.width = width
        self.height = height    
        self.n_citizens = n_citizens
        self.n_races = n_races
        self.sim_threshold = sim_threshold
        
        for i in range(n_citizens):
            a = Citizen(i, self)
            coor = loc[i]
            self.schedule.add(a)
            self.grid.place_agent(a, coor)
        
        for a in self.schedule.agents:
            a.look_neighbor()
            a.similarity()
            a.decide()

        self.grid.occupied = list(set(self.grid.grids) - set(self.grid.empties))

    def step(self):
        self.schedule.step()
        self.n_steps += 1
        
    def plot(self):
        cells = np.zeros((self.width, self.height))
        for contents in self.grid.coord_iter():
            content, x, y = contents
            if content != None:
                cells[x][y] = content.race
        plt.imshow(cells)
        plt.axis("off")

model = City(50, 50, 2000, 2, 0.6)
```

```python
model = City(50, 50, 2000, 2, 0.6)
plt.figure(figsize=(10,5))
for i in range(8):
    plt.subplot(2,4,i+1)
    model.plot()
    plt.title("{} Update(s)".format(model.n_steps))
    model.step()
```
